# Exercício 4

Exercício 4

```python
n = 100
columns = ['Brand '+str(number) for number in range(1,5)]
date = pd.date_range(start = '2021-1-1', periods = n)
data = np.random.randint(low = 1, high = [100, 330, 45, 77], size = (n, 4))
df = pd.DataFrame(data,columns = columns, index = date)
```

```python
df
```

```python
brand1 = df['Brand 1']
```

```python
plt.figure(figsize = (20, 5))
#plt.plot(df.index, df['Brand 1'], color = 'black') resposta do exercício
plt.plot(date, brand1, color = 'black')
#foi colocado no argumento de x 'df.index', mas poderia ter sido 'data'
#foi selecionado a coluna dentro do argumento, mas a opção anterior atribuia a uma variável a coluna e utilizaria a variável
plt.title("Brand 1")
plt.xlabel("Data")
plt.ylabel("Values")
plt.show()
```

# Exercício 9


Exercício 9

```python
x = np.arange(1,11)
y1 = x**2
y2 = x**3
y3 = x**4
```

```python
plt.scatter(x, y1, marker = 'o')
plt.scatter(x, y2, marker = '^')
plt.scatter(x, y3, marker = '*')
plt.title('Values')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True, linestyle = '--')
plt.legend(['y1','y2', 'y3'], title = "Legend")
plt.show()
```
