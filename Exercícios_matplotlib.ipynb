```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Exercício 1

```python
x = np.arange(-5.0, 5.0, 0.01)
y = np.sin(np.pi*x)
```

```python
plt.plot(x, y)
plt.title("título")
plt.xlabel("eixo x")
plt.ylabel("eixo y")
plt.show()
```

Exerxicio 2

```python
x = np.arange(20)
y = np.random.randint(1, 100, 20)
```

```python
plt.plot(x, y, marker = 'o', color = 'black')
plt.title("Título")
plt.xlabel("eixo x")
plt.ylabel("eixo y")
plt.show()
```

Exercício 4

```python
n = 100
columns = ['Brand '+str(number) for number in range(1,5)]
date = pd.date_range(start = '2021-1-1', periods = n)
data = np.random.randint(low = 1, high = [100, 330, 45, 77], size = (n, 4))
df = pd.DataFrame(data,columns = columns, index = date)
```

```python
df
```

```python
brand1 = df['Brand 1']
```

```python
plt.figure(figsize = (20, 5))
#plt.plot(df.index, df['Brand 1'], color = 'black') resposta do exercício
plt.plot(date, brand1, color = 'black')
#foi colocado no argumento de x 'df.index', mas poderia ter sido 'data'
#foi selecionado a coluna dentro do argumento, mas a opção anterior atribuia a uma variável a coluna e utilizaria a variável
plt.title("Brand 1")
plt.xlabel("Data")
plt.ylabel("Values")
plt.show()
```
