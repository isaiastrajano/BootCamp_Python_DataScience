```python
np.random.seed(7)
dados = np.random.randint(low = 100, high = [250, 300, 500, 1000], size = (12, 4))
columns = ["A", "B", "C", "D"]
index = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]
```

```python
data = pd.DataFrame(dados, columns = columns, index = index)
```

```python
data
```

```python
plt.figure(figsize = (15, 5))
plt.plot(index,data["A"])
plt.plot(index,data["B"])
plt.plot(index,data["C"])
plt.plot(index,data["D"])
plt.title("Valores Over Time")
plt.xlabel("Years")
plt.ylabel("X-Axis")
plt.ylim(80, 1000)
plt.grid(True, linestyle = "--", alpha = 0.5)
plt.legend(columns, loc = 'upper left')
plt.show()
```
